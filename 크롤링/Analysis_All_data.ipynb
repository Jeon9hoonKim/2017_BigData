# 1. KDA analysis
## 0) DataLoad

```python
# get data file names
path =r'C:\Users\waf\Documents\00_DGIST 수업\03_02\Big data\프로젝트\2017_BigData\데이터\Data\kda_analysis'
# filenames = glob.glob(path + "/*.csv")
filenames = glob.glob(path + "/*.csv")
filenames
dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename, encoding="utf8", engine='python'))
# Concatenate all data into one DataFrame
kda_data = pd.concat(dfs, ignore_index=True)
kda_data['Death_fake'] = kda_data['Death']
kda_data['Death_fake'] = kda_data['Death_fake'].replace(0, 1)
kda_data['KDA'] = (kda_data['Kill'] + kda_data['Assist'])/ kda_data['Death_fake']
kda_data['KDA2'] = kda_data['Kill'] + 0.5*kda_data['Assist'] - 1.2*kda_data['Death']
kda_data = kda_data.drop('Death_fake', 'columns')
kda_data
```

## 1) simple sum comparison

### 1-1) Kill Sum

```python
more_kill_win = 0
more_kill_lose = 0
not_more_kill = 0
total = 0
tmp = []
index = 0
for i in kda_data.groupby(['MatchNumber','GameResult'])['Kill'].sum():
    tmp.append(i)
    index += 1
    if index == 2:
        total += 1
        if tmp[0] > tmp[1]:
            more_kill_win += 1
        elif tmp[0] < tmp[1]:
            more_kill_lose += 1
        else:
            not_more_kill += 1
        tmp = []
        index = 0
# print(more_kill_win/total)
print("more win  :",more_kill_win/total)
print("more lose :",more_kill_lose/total)
print("same      :",not_more_kill/total)
```
