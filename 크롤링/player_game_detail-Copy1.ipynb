```python
KeystoneMastery = []
KeystoneMastery_raw = PlayerData.find_all('div',{'class':'KeystoneMastery'})
for line in KeystoneMastery_raw:
    line = str(line)
    line = line[line.find("#ffc659'&gt;"):line.find('&lt;/b&gt;')].replace("#ffc659'&gt;",'').replace('&lt;/b&gt;','')
    KeystoneMastery.append(line)
print(KeystoneMastery)
print(len(KeystoneMastery))
```

```python
kill = []
death = []
assist = []
kda_ratio = []
kill_raw = PlayerData.find_all('span',{'class':'Kill'})
death_raw = PlayerData.find_all('span',{'class':'Death'})
assist_raw = PlayerData.find_all('span',{'class':'Assist'})
kda_ratio_raw = PlayerData.find_all('span',{'class':'KDARatio'})

for line in kill_raw:
    line = str(line)
    line = line[line.find('"Kill">'):line.find('</span>')].replace('"Kill">','').replace('</span>','')
    if line not in ['더블킬', '트리플킬', '쿼드라킬', '펜타킬']:
        kill.append(line)

for line in death_raw:
    line = str(line)
    line = line[line.find('"Death">'):line.find('</span>')].replace('"Death">','').replace('</span>','')
    death.append(line)

for line in assist_raw:
    line = str(line)
    line = line[line.find('"Assist">'):line.find('</span>')].replace('"Assist">','').replace('</span>','')
    assist.append(line)
    
for line in kda_ratio_raw:
    line = str(line)
    line = line[line.find('">'):line.find('</span>')].replace('">','').replace('</span>','')
    kda_ratio.append(line)

print(kill)
print(len(kill))
print(death)
print(len(death))
print(assist)
print(len(assist))
print(kda_ratio)
print(len(kda_ratio))
```

```python
level = []
cs = []
cs_per_min = []
ward = []
kill_participation = []
level_raw = PlayerData.find_all('div',{'class':'Level'})
cs_raw = PlayerData.find_all('div',{'class':'CS'})
ward_raw = PlayerData.find_all('div',{'class':'Ward'})
kill_participation_raw = PlayerData.find_all('div',{'class':'CKRate tip'})

for line in level_raw:
    line = str(line)
    line = line[line.find('레벨'):line.find('</div>')].replace('레벨','').replace('\n','').replace('\t','')
    if line != '':
        level.append(line)
    
for line in cs_raw:
    line = str(line)
    line1 = line[line.find('개">'):line.find(' (')].replace('개">','').replace(' (','')
    line2 = line[line.find('('):line.find(')')].replace('(','').replace(')','')
    if line1 != '':
        cs.append(line1)
    if line2 != '':
        cs_per_min.append(line2)
    
for line in ward_raw:
    line = str(line)
    line = line[line.find('"wards vision">'):line.find('</span>')].replace('"wards vision">','').replace('</span>','')
    ward.append(line)
    
for line in kill_participation_raw:
    line = str(line)
    line = line[line.find('킬관여 '):line.find('%')+1].replace('킬관여 ','')
    kill_participation.append(line)

print(level)
print(len(level))
print(cs)
print(len(cs))
print(cs_per_min)
print(len(cs_per_min))
print(ward)
print(len(ward))
print(kill_participation)
print(len(kill_participation))
```
