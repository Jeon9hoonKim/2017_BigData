## 필요한 라이브러리 import

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import glob
```

## csv 파일 불러오기 및 KDA 계산

```python
# get data file names
path =r'C:\Users\user\Documents\Python Workspace\mid 3\progamer 긁기\data\SGvLG' # csv가 저장된 경로 지정
# filenames = glob.glob(path + "/*.csv")
filenames = glob.glob(path + "/*.csv") # 
filenames # 
dfs = [] # csv를 합칠 리스트 지정
for filename in filenames:
    dfs.append(pd.read_csv(filename, engine='python')) # csv 합치기
# Concatenate all data into one DataFrame
kda_data = pd.concat(dfs, ignore_index=True) # 
kda_data['Death_fake'] = kda_data['Death'] # KDA 계산을 위해 'Death' coulumn 복제
kda_data['Death_fake'] = kda_data['Death_fake'].replace(0, 1) # 0으로 나누면 inf가 되므로 1로 변경
kda_data['KDA'] = (kda_data['Kill'] + kda_data['Assist'])/ kda_data['Death_fake'] # KDA 계산
kda_data['KDA2'] = kda_data['Kill'] + 0.5*kda_data['Assist'] - 1.2*kda_data['Death'] # KDA2 계산
kda_data = kda_data.drop('Death_fake', 'columns') # 계산용 Death_fake 지우기
kda_data # DataFrame 확인
```

## 프로게이머 이름 수정
 프로게이머 한 명이 여러 계정을 사용하는 경우도 있고, 게임의 닉네임이 대중적으로 알려져있는 이름과 다른 경우도 있어 수정하였다.

```python
kda_data['Nickname'] = kda_data['Nickname'].replace(['shirushi','shirushi2','삼성갤럭시 하루','Dear Brave'], 'Haru')
kda_data['Nickname'] = kda_data['Nickname'].replace('삼성갤럭시 cuvee', 'CuVee')
kda_data['Nickname'] = kda_data['Nickname'].replace(['삼성갤럭시크라운','삼성갤럭시 Crown'], 'Crown')
kda_data['Nickname'] = kda_data['Nickname'].replace(['RRRuleRRR','삼성갤럭시 Ru1er'], 'Ruler')
kda_data['Nickname'] = kda_data['Nickname'].replace(['1521819del','삼성갤럭시CoreJJ','CoreJangJeon'], 'CoreJJ')

kda_data['Nickname'] = kda_data['Nickname'].replace(['smeb00','춘봉박'], 'Smeb')
kda_data['Nickname'] = kda_data['Nickname'].replace(['좋은날','열심히할게요오','1110551del'], 'Score')
kda_data['Nickname'] = kda_data['Nickname'].replace(['케이티 PawN','똥좀싸겟다','PawN God'], 'Pawn')
kda_data['Nickname'] = kda_data['Nickname'].replace(['deft123456','KT DEFT','열심히좀해라','Deft1'], 'Deft')
kda_data['Nickname'] = kda_data['Nickname'].replace(['KT MATA','시어니보고싶움','MataMataMata'], 'Mata')

kda_data['Nickname'] = kda_data['Nickname'].replace(['SKT T1 Untara','뒷텔와드점','몸이아파아똑딱해'], 'Untara')
kda_data['Nickname'] = kda_data['Nickname'].replace(['SKT T1 BIank','멜리크리스마스','갱만가드림'], 'Blank')
kda_data['Nickname'] = kda_data['Nickname'].replace(['SKT T1 Faker','Hide on bush'], 'Faker')
kda_data['Nickname'] = kda_data['Nickname'].replace(['SKT T1 Bang'], 'Bang')
kda_data['Nickname'] = kda_data['Nickname'].replace(['SKT T1 WoIf'], 'Wolf')

kda_data['Nickname'] = kda_data['Nickname'].replace(['Co2dplay', '피넛사냥개2'], 'Khan')
kda_data['Nickname'] = kda_data['Nickname'].replace(['PeanuToT','Jon Sn0w'], 'Peanut')
kda_data['Nickname'] = kda_data['Nickname'].replace(['8강따리미드', '피넛사냥개1'], 'Wolf')
kda_data['Nickname'] = kda_data['Nickname'].replace(['학식먹을나이', '구직자 김종인', '이길수없어'], 'PraY')
kda_data['Nickname'] = kda_data['Nickname'].replace(['강고릴'], 'GorillA')
```

## 분석을 위해 kda_data를 Nickname으로 groupby 해서 필요한 자료만 얻는다.

```python
match = kda_data.groupby('Nickname')[['Kill','Death','Assist','Damage','Buy_Pward','Build_ward','Destroy_ward','CS','Baron','Dragon','Tower','KDA','KDA2']].mean()
match.head()
```

## 관심의 대상이 되는 프로게이머만 뽑아낸다.

```python
match_cut = match.loc[['Khan', 'Peanut', 'Wolf', 'PraY', 'GorillA', 'CuVee','Haru','Crown','Ruler','CoreJJ'],:]
match_cut
```

## csv로 저장
