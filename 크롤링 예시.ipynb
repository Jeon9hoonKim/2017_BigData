## 크롤링하기

```python
import requests
 
# GET
url = 'http://www.op.gg/summoner/userName=tar2an'
resp = requests.post(url)
print(resp.text)
```

## 크롤링 후 텍스트 제거

```python
import requests
import bs4
url = 'http://www.op.gg/summoner/userName=tar2an'
resp = requests.get(url)
bs = bs4.BeautifulSoup(resp.text, 'html.parser')
print(bs.get_text())
```
